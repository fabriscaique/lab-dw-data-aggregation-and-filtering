This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

```python
file = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
```

```python
draft = pd.read_csv(file)
```

```python
df = pd.DataFrame(draft)
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.info()
```

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
df2 = df[(df['Total Claim Amount'] > 1000) & (df['Response'] == 'Yes')]
df2.reset_index(drop=True, inplace=True)
df2
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
df3 = df[df['Response'] == 'Yes']
df3.reset_index(drop=True, inplace=True)
```

```python
table = df3.pivot_table(
    index = 'Policy Type',
    columns = 'Gender',
    values = 'Total Claim Amount',
    aggfunc = 'mean'
)

table.round(2)
# The average claims vary by policy type, indicating that certain policies may attract different demographic behaviors 
# females tend to file higher claims in both Corporate Auto and Special Auto policies
# males have slightly higher claims in Personal Auto.

# I won't do it, but I bet if we ran a real correlation test, differences would not be significant across any parameters, though
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
df3 = df.groupby(['Gender', 'Education'])['Customer Lifetime Value'].agg(['max', 'min', 'median']).reset_index()
df3

# Again, customer lifetime value tends to be higher in lower education demographics
# This suggests that customers with lower education might be target for specific campaigns
```
